This is due to pressence of extremely large outliers that heavily skewed the data to the right, making estimation use the mean very inaccurate. In this case, I believe that using the median of $284.0 would more accurately represent how much customers tend to spend in a single visit

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv("orders.csv")
print(df.shape)
df.head(17)
```

```python
order_sum = df['order_value'].sum()
print(order_sum/5000)
order_max = df['order_value'].max()
print(order_max)
df_order = df.set_index('order_value')
df_order.loc[704000]
#Very large outliers
#When the order value is high, customers tend to pay using credit cards
#All of the orders with values of 704000 are by the same user(607), from the same shop(42), and occured at the same time of the day(4:00:00).
```

```python
df['order_value'].plot.box(vert = False)
#Heavily skewed to the right due to the extremely large outliers
#The median should be used instead of the mean when describing the spread of the data
```

```python
df['order_value'].plot.box(vert = False, showfliers = False)
#When disregarding the outliers, the middle 50% of the order values are between approximately 160 and 390
```

```python
order_value_median = df['order_value'].median()
print(order_value_median)
dif = (df['order_value'] - order_value_median).sum()
print(dif)
df['value per item'] = df['order_value']/df['total_items']
df.head(20)
```

```python
df.groupby('payment_method').sum()
#This shows that the majority of items bought were payed for using credit cards
```

```python
df2 = df[df.user_id != 607]
df2.sort_values(by = 'total_items', ascending = False).head(20)
#This shows that if we remove the oddity(user 607), the most pairs of shoes a customer bought at a single time is 8.
df2['total_items'].plot.kde()
#This shows that if we remove the oddity, most customers buy 1 or 2 pairs of shoes in a single visit
```

```python
df3 = df2[df2.order_value < 1000]
df3['order_value'].plot.kde()
#This shows that most of the orders have total values between 80 and 500
```
